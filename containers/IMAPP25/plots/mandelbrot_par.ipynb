## **Code I used for plotting** ##

(Not that you need it, but anyway)

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Modify the file path and file name
df = pd.read_csv("results.txt", sep=r'\s+', skiprows=1, names=['Grain', 'Time [ms]', 'Tasks'])
df
```

```python
plt.figure(figsize=(14, 4), dpi=300)
plt.subplot(1, 2, 1)
plt.scatter(df['Grain'], df['Time [ms]'], s = 15)
plt.grid()
plt.xlabel('Grain')
plt.ylabel('Time [ms]')
plt.title('Grain vs execution time [ms]')
plt.subplot(1, 2, 2)
plt.scatter(df['Grain'], df['Tasks'], s = 15)
plt.grid()
plt.xlabel('Grain')
plt.ylabel('Number of tasks')
plt.title('Grain vs number of tasks')
plt.savefig("grain_time_tasks.png", bbox_inches='tight');
```

```python
plt.figure(figsize=(14, 4), dpi=300)
plt.subplot(1, 2, 1)
plt.scatter(df['Grain'], df['Time [ms]'], s = 15)
plt.xscale('log')
plt.grid()
plt.xlabel('Grain (Log)')
plt.ylabel('Time [ms]')
plt.title('Grain vs execution time [ms]')
plt.subplot(1, 2, 2)
plt.scatter(df['Grain'], df['Tasks'], s = 15)
plt.xscale('log')
plt.grid()
plt.xlabel('Grain (Log)')
plt.ylabel('Number of tasks')
plt.title('Grain vs number of tasks')
plt.savefig("log_grain_time_tasks.png", bbox_inches='tight');
```

```python
plt.figure(figsize=(7, 4), dpi=300)
plt.scatter(df['Grain'], df['Tasks'], s = 15)
plt.xscale('log')
plt.yscale('log')
plt.grid()
plt.xlabel('Grain (Log)')
plt.ylabel('Number of tasks (Log)')
plt.title('Grain vs number of tasks')
plt.savefig("log_grain_log_time_tasks.png", bbox_inches='tight');
```
